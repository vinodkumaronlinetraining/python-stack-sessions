```python
class User:
    def __init__(self, name):
        self.name = name

    def __str__(self): # called whenever we use print statements with object
        return f"User: {self.name}"
    
    def __repr__(self): 
        return f"User(name='{self.name}')"
    
user1 = User("Vinod")
user2 = User("manoj")
print(user1)
user1
user2
```
